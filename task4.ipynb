```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_img(img):
    plt.figure(figsize=(10,6))
    plt.imshow(img,cmap='gray')
    plt.axis("off")
```

```python
image = cv2.imread("img.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plot_img(image)
```

```python

img_gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plot_img(img_gray)
```

```python
inverted_img = 255 - img_gray
plot_img(inverted_img)
```

```python
blurred_img = cv2.GaussianBlur(inverted_img,(25,25),0)
plot_img(blurred_img)
```

```python
inverted_blur = 255 - blurred_img
skitch = cv2.divide(img_gray,inverted_blur,scale=256.0)
plot_img(skitch)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(skitch,cmap='gray')
plt.title("Skecth")
plt.axis("off")
```
